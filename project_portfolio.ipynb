#Data Cleansing

````python
df_clean = df.copy()
````

````python
print(f"Missing values before cleaning: {df_clean.isnull().sum().sum()}")
````

````python
# 1. Convert TotalCharges to numeric if it's stored as string
if 'Total Charges' in df_clean.columns:
    df_clean['Total Charges'] = pd.to_numeric(df_clean['Total Charges'], errors='coerce')
````

````python
 # 2. Handle missing values in Total Charges (common issue)
if 'Total Charges' in df_clean.columns and df_clean['Total Charges'].isnull().sum() > 0:
    # Fill missing TotalCharges with 0 for new customers or calculate from tenure and monthly charges
    if 'Tenure Months' in df_clean.columns and 'Monthly Charges' in df_clean.columns:
        df_clean['Total Charges'] = df_clean['Total Charges'].fillna(
            df_clean['Tenure Months'] * df_clean['Monthly Charges']
        )
    else:
        df_clean['Total Charges'] = df_clean['Total Charges'].fillna(0)
````

````python
# 3. Standardize Yes/No columns
yes_no_cols = []
for col in df_clean.columns:
    if df_clean[col].dtype == 'object':
        unique_vals = df_clean[col].unique()
        if len(unique_vals) == 2 and set(unique_vals) == {'Yes', 'No'}:
            yes_no_cols.append(col)
print(f"Yes/No columns found: {yes_no_cols}")
````

````python
 # 4. Handle categorical variables - ensure consistent formatting
categorical_cols = df_clean.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df_clean[col] = df_clean[col].str.strip()  # Remove whitespace
````

````python
# 3. Handle Churn Reason column (CRITICAL DECISION)
churn_reason_cols = [col for col in df_clean.columns if 'reason' in col.lower() or 'churn' in col.lower()]

print(f"\nChurn-related columns found: {churn_reason_cols}")
````

````python
# BUSINESS DECISION: Keep churn reason column for churned customers analysis
# This is valuable for understanding WHY customers churn
for col in churn_reason_cols:
    if col in df_clean.columns and col.lower() != 'churn':
        missing_count = df_clean[col].isnull().sum()
        churned_count = (df_clean['Churn'] == 'Yes').sum() if 'Churn' in df_clean.columns else 0

        print(f"\n🔍 CHURN REASON ANALYSIS:")
        print(f"  Column: {col}")
        print(f"  Missing values: {missing_count}")
        print(f"  Churned customers: {churned_count}")
        print(f"  Expected missing: {len(df_clean) - churned_count} (non-churned customers)")

````

````python
# Verify that missing values correspond to non-churned customers
if 'Churn' in df_clean.columns:
    non_churned_with_reason = df_clean[(df_clean['Churn'] == 'No') & (df_clean[col].notna())].shape[0]
    churned_without_reason = df_clean[(df_clean['Churn'] == 'Yes') & (df_clean[col].isna())].shape[0]

    print(f"  Non-churned customers with reason: {non_churned_with_reason}")
    print(f"  Churned customers without reason: {churned_without_reason}")

    # DECISION: Keep the column, missing values are expected for retained customers
    if missing_count > 0:
        print(f"  ✅ KEEPING {col} - Missing values are expected for retained customers")
        print(f"  ✅ Will use for churn reason analysis among churned customers only")

        # Optional: Fill missing values with "Not Applicable" for clarity
        # df_clean[col] = df_clean[col].fillna('Not Applicable - Customer Retained')
````

````python
# Ensure 'Churn Label' and 'Churn Reason' columns exist
if 'Churn Label' in df_clean.columns and 'Churn Reason' in df_clean.columns:
    # Identify rows where 'Churn Label' is 'No' and 'Churn Reason' is null
    condition = (df_clean['Churn Label'] == 'No') & (df_clean['Churn Reason'].isnull())

    # Fill the 'Churn Reason' with 'No Churn' for these specific rows
    df_clean.loc[condition, 'Churn Reason'] = 'No Churn'

    print("\nFilling null 'Churn Reason' with 'No Churn' for 'Churn Label' == 'No':")
    # Verify the changes - check null counts in 'Churn Reason' again
    print(df_clean.isna().sum()['Churn Reason'])

    # Optionally, check a sample of rows where this change was made
    # print("\nSample rows after filling nulls:")
    # display(df_clean[condition].head())
else:
    print("Columns 'Churn Label' or 'Churn Reason' not found in the DataFrame.")
````

````python
df_clean.columns = df_clean.columns.str.replace(' ', '_')
df_clean.columns
````

#Feature Engineering

````python
# 1. Tenure Categories (Year)
if 'Tenure_Months' in df_clean.columns:
    df_clean['Tenure_Category'] = pd.cut(
        df_clean['Tenure_Months'],
        bins=[0, 12, 24, 48, 72],
        labels=['0-1 Year', '1-2 Years', '2-4 Years', '4-6 Years'],
        right=True
    )
    df_clean.loc[df_clean['Tenure_Months'] == 0, 'Tenure_Category'] = '0-1 Year'

print("\nAdding 'Tenure_Category' column:")
print(df_clean[['Tenure_Months', 'Tenure_Category']].head())
print("\nValue Counts for 'Tenure_Category':")
print(df_clean['Tenure_Category'].value_counts())
````

````python
# 2. Monthly Charges Categories
if 'Monthly_Charges' in df_clean.columns:
    df_clean['monthly_charges_category'] = pd.cut(
          df_clean['Monthly_Charges'],
        bins=[0, 35, 50, 65, 80, float('inf')],
        labels=['Low ($0-35)', 'Medium-Low ($35-50)', 'Medium ($50-65)', 'Medium-High ($65-80)', 'High ($80+)']
    )
````

````python
# 3. Service Usage Score (count of services used)
service_cols = [col for col in df_clean.columns if col in [
      'Phone_Service', 'Multiple_Lines', 'Internet_Service', 'Online_Security',
    'Online_Backup', 'Device_Protection', 'Tech_Support', 'Streaming_TV', 'Streaming_Movies'
]]
if service_cols:
    df_clean['service_usage_score'] = 0
    for col in service_cols:
        if col in df_clean.columns:
            df_clean['service_usage_score'] += (df_clean[col] == 'Yes').astype(int)
````

````python
# 4. Customer Value Score (based on tenure and charges)
if 'Tenure_Months' in df_clean.columns and 'Monthly_Charges' in df_clean.columns:
    df_clean['customer_value_score'] = (
        (df_clean['Tenure_Months'] / df_clean['Tenure_Months'].max()) * 0.4 +
        (df_clean['Monthly_Charges'] / df_clean['Monthly_Charges'].max()) * 0.6
    ) * 100
````

````python
# 5. Risk Categories
if 'Contract' in df_clean.columns and 'Payment_Method' in df_clean.columns:
    df_clean['risk_category'] = 'Medium'
    df_clean.loc[
        (df_clean['Contract'] == 'Month-to-month') &
        (df_clean['Payment_Method'] == 'Electronic check'),
        'risk_category'
    ] = 'High'
    df_clean.loc[
        (df_clean['Contract'].isin(['One year', 'Two year'])) &
        (df_clean['Payment_Method'] != 'Electronic check'),
        'risk_category'
    ] = 'Low'
````

````python
# 6. Average Monthly Charges per Service
if 'Monthly_Charges' in df_clean.columns and 'service_usage_score' in df_clean.columns:
    df_clean['avg_charge_per_service'] = np.where(
          df_clean['service_usage_score'] > 0,
        df_clean['Monthly_Charges'] / df_clean['service_usage_score'],
        df_clean['Monthly_Charges']
    )
````

````python
 # 7. Senior Customer Flag
if 'Senior_Citizen' in df_clean.columns:
    df_clean['is_senior'] = df_clean['Senior_Citizen'].map({'Yes': 1, 'No': 0})
````

#Key Insights

````python
# prompt: buat skrip untuk mengisi nilai null pada kolom churn_Category. sesuaikan dengan churn_label

# Isi nilai null pada Churn_Category berdasarkan Churn_Label
# Jika Churn_Label adalah 'No', maka Churn_Category seharusnya null (atau 'No Churn')
# Jika Churn_Label adalah 'Yes', maka Churn_Category seharusnya terisi

if 'Churn_Label' in df_clean.columns and 'Churn_Category' in df_clean.columns:
    # Filter baris di mana Churn_Label adalah 'No' dan Churn_Category adalah null
    condition_fill_no = (df_clean['Churn_Label'] == 'No') & (df_clean['Churn_Category'].isnull())
    # Isi Churn_Category dengan nilai 'No Churn' untuk baris yang memenuhi kondisi
    df_clean.loc[condition_fill_no, 'Churn_Category'] = 'No Churn'

    # Kita juga bisa memeriksa apakah ada baris Churn_Label 'Yes' yang Churn_Categorynya null
    # Jika ada, ini mungkin data yang perlu diperbaiki atau diinvestigasi lebih lanjut.
    # Namun, berdasar deskripsi, fokusnya adalah mengisi null saat Churn_Label 'No'.

    print("\nMengisi null pada 'Churn_Category' berdasarkan 'Churn_Label':")
    print(f"Jumlah null di 'Churn_Category' sebelum diisi: {df_clean['Churn_Category'].isnull().sum()}")

    # Verifikasi pengisian
    # Cek jumlah null di Churn_Category setelah diisi
    print(f"Jumlah null di 'Churn_Category' setelah diisi: {df_clean['Churn_Category'].isnull().sum()}")

    # Tampilkan contoh baris yang Churn_Labelnya 'No'
    print("\nContoh baris dengan 'Churn_Label' == 'No' dan 'Churn_Category' terisi:")
    display(df_clean[df_clean['Churn_Label'] == 'No'][['Churn_Label', 'Churn_Category']].head())

    # Cek apakah ada baris 'Churn_Label' == 'Yes' yang 'Churn_Category' nya masih null
    if df_clean[(df_clean['Churn_Label'] == 'Yes') & (df_clean['Churn_Category'].isnull())].shape[0] > 0:
        print("\nPERINGATAN: Ada baris dengan 'Churn_Label' == 'Yes' yang 'Churn_Category'-nya masih null.")
        print("Anda mungkin perlu menangani baris ini secara terpisah (misal: mengisi dengan 'Unknown' atau 'Other').")
else:
    print("Kolom 'Churn_Label' atau 'Churn_Category' tidak ditemukan di DataFrame.")

```
````

````python
# Churn Analysis
churn_col = 'Churn_Label' if 'Churn_Label' in df_clean.columns else 'Churn'
if churn_col in df_clean.columns:
    churn_rate = (df_clean[churn_col] == 'Yes').mean()
    print(f"\n📊 OVERALL CHURN METRICS:")
    print(f"Total Customers: {len(df_clean):,}")
    print(f"Churned Customers: {(df_clean[churn_col] == 'Yes').sum():,}")
    print(f"Overall Churn Rate: {churn_rate:.2%}")

    # Churn Reason Analysis (if available)
    if 'Churn_Reason' in df_clean.columns:
        churned_customers = df_clean[df_clean[churn_col] == 'Yes']
        churn_reasons = churned_customers['Churn_Reason'].value_counts()

        print(f"\n🎯 CHURN REASON ANALYSIS:")
        print(f"Top reasons for customer churn:")
        for i, (reason, count) in enumerate(churn_reasons.head(10).items(), 1):
            percentage = (count / len(churned_customers)) * 100
            print(f"  {i}. {reason}: {count} customers ({percentage:.1f}%)")

        # Group churn reasons into categories for strategic analysis
        churn_categories = {}
        for reason in churn_reasons.index:
            reason_lower = reason.lower()
            if any(word in reason_lower for word in ['price', 'expensive', 'cost', 'bill']):
                churn_categories[reason] = 'Price/Cost Issues'
            elif any(word in reason_lower for word in ['competitor', 'better', 'offer']):
                churn_categories[reason] = 'Competitive Offer'
            elif any(word in reason_lower for word in ['service', 'quality', 'poor', 'bad']):
                churn_categories[reason] = 'Service Quality'
            elif any(word in reason_lower for word in ['support', 'customer', 'help']):
                churn_categories[reason] = 'Customer Support'
            elif any(word in reason_lower for word in ['moved', 'relocation', 'location']):
                churn_categories[reason] = 'Relocation'
            elif any(word in reason_lower for word in ['no', 'churn']):
                churn_categories[reason] = 'Active Customer'
            else:
                churn_categories[reason] = 'Other'

        # Add churn category column
        df_clean['churn_category'] = df_clean['Churn_Reason'].map(churn_categories)

        # Add No Churn to Customer who choose No
        if 'Churn_Label' in df_clean.columns and 'churn_category' in df_clean.columns:
          condition_fill_no = (df_clean['Churn_Label'] == 'No') & (df_clean['churn_category'].isnull())
          df_clean.loc[condition_fill_no, 'churn_category'] = 'No Churn'

        # Analyze churn categories
        category_analysis = df_clean[df_clean[churn_col] == 'Yes']['churn_category'].value_counts()
        print(f"\n📋 CHURN CATEGORIES:")
        for category, count in category_analysis.items():
            if category != 'Active Customer':
                percentage = (count / len(churned_customers)) * 100
                print(f"  {category}: {count} customers ({percentage:.1f}%)")

    # Churn by key factors
    print(f"\n📈 CHURN BY KEY FACTORS:")

    # By Contract Type
    if 'Contract' in df_clean.columns:
        contract_churn = df_clean.groupby('Contract')[churn_col].apply(
            lambda x: (x == 'Yes').mean()
        ).sort_values(ascending=False)
        print(f"\nChurn by Contract Type:")
        for contract, rate in contract_churn.items():
            print(f"  {contract}: {rate:.2%}")

    # By Tenure Category
    if 'Tenure_Category' in df_clean.columns:
        tenure_churn = df_clean.groupby('Tenure_Category')[churn_col].apply(
            lambda x: (x == 'Yes').mean()
        ).sort_values(ascending=False)
        print(f"\nChurn by Tenure Category:")
        for tenure, rate in tenure_churn.items():
            print(f"  {tenure}: {rate:.2%}")

    # By Payment Method
    if 'Payment_Method' in df_clean.columns:
        payment_churn = df_clean.groupby('Payment_Method')[churn_col].apply(
            lambda x: (x == 'Yes').mean()
        ).sort_values(ascending=False)
        print(f"\nChurn by Payment Method:")
        for payment, rate in payment_churn.items():
            print(f"  {payment}: {rate:.2%}")

    # By Risk Category
    if 'risk_category' in df_clean.columns:
        risk_churn = df_clean.groupby('risk_category')[churn_col].apply(
            lambda x: (x == 'Yes').mean()
        ).sort_values(ascending=False)
        print(f"\nChurn by Risk Category:")
        for risk, rate in risk_churn.items():
            print(f"  {risk}: {rate:.2%}")
````

````python
# Revenue Analysis
if 'Monthly_Charges' in df_clean.columns and churn_col in df_clean.columns:
    print(f"\n💰 REVENUE IMPACT ANALYSIS:")
    churned_revenue = df_clean[df_clean[churn_col] == 'Yes']['Monthly_Charges'].sum()
    retained_revenue = df_clean[df_clean[churn_col] == 'No']['Monthly_Charges'].sum()
    total_revenue = churned_revenue + retained_revenue

    print(f"Monthly Revenue Lost to Churn: ${churned_revenue:,.2f}")
    print(f"Monthly Revenue Retained: ${retained_revenue:,.2f}")
    print(f"Revenue at Risk: {(churned_revenue/total_revenue):.2%}")

    # Annual impact
    annual_lost_revenue = churned_revenue * 12
    print(f"Estimated Annual Revenue Lost: ${annual_lost_revenue:,.2f}")

    # Revenue impact by churn category
    if 'churn_category' in df_clean.columns:
        print(f"\n💸 REVENUE IMPACT BY CHURN CATEGORY:")
        churn_revenue_by_category = df_clean[df_clean[churn_col] == 'Yes'].groupby('churn_category')['Monthly_Charges'].sum().sort_values(ascending=False)
        for category, revenue in churn_revenue_by_category.items():
            if category != 'Active Customer':
                annual_impact = revenue * 12
                print(f"  {category}: ${revenue:,.2f}/month (${annual_impact:,.2f}/year)")
````

````python
# Customer Segmentation Analysis
if 'customer_value_score' in df_clean.columns and churn_col in df_clean.columns:
    print(f"\n🎯 CUSTOMER SEGMENTATION INSIGHTS:")

    # High-value customers at risk
    high_value_threshold = df_clean['customer_value_score'].quantile(0.8)
    high_value_customers = df_clean[df_clean['customer_value_score'] >= high_value_threshold]
    high_value_churn_rate = (high_value_customers[churn_col] == 'Yes').mean()

    print(f"High-Value Customer Churn Rate: {high_value_churn_rate:.2%}")
    print(f"High-Value Customers at Risk: {(high_value_customers[churn_col] == 'Yes').sum():,}")

    # Service usage impact
    if 'service_usage_score' in df_clean.columns:
        service_churn = df_clean.groupby('service_usage_score')[churn_col].apply(
            lambda x: (x == 'Yes').mean()
        )
        print(f"\nChurn by Service Usage Score:")
        for score, rate in service_churn.items():
            print(f"  {score} services: {rate:.2%}")
````

#Strategic Recommendation

````python
recommendations = []

# Churn reason-based recommendations
if 'churn_category' in df_clean.columns and churn_col in df_clean.columns:
    category_analysis = df_clean[df_clean[churn_col] == 'Yes']['churn_category'].value_counts()

    # Price/Cost Issues
    if 'Price/Cost Issues' in category_analysis.index:
        price_churn_pct = (category_analysis['Price/Cost Issues'] / len(df_clean[df_clean[churn_col] == 'Yes'])) * 100
        if price_churn_pct > 20:
            recommendations.append({
                'priority': 'HIGH',
                'area': 'Pricing Strategy',
                'recommendation': 'Implement competitive pricing and value communication program',
                'details': f'{price_churn_pct:.1f}% of churn is due to price/cost issues. Develop retention pricing tiers and value proposition messaging.',
                'expected_impact': '20-30% reduction in price-related churn'
            })

    # Competitive Offer
    if 'Competitive Offer' in category_analysis.index:
        comp_churn_pct = (category_analysis['Competitive Offer'] / len(df_clean[df_clean[churn_col] == 'Yes'])) * 100
        if comp_churn_pct > 15:
            recommendations.append({
                'priority': 'HIGH',
                'area': 'Competitive Intelligence',
                'recommendation': 'Enhance competitive monitoring and counter-offer programs',
                'details': f'{comp_churn_pct:.1f}% of churn is due to competitive offers. Implement win-back campaigns and competitive matching.',
                'expected_impact': '15-25% reduction in competitive churn'
            })

    # Service Quality
    if 'Service Quality' in category_analysis.index:
        quality_churn_pct = (category_analysis['Service Quality'] / len(df_clean[df_clean[churn_col] == 'Yes'])) * 100
        if quality_churn_pct > 10:
            recommendations.append({
                'priority': 'HIGH',
                'area': 'Service Quality',
                'recommendation': 'Implement comprehensive service quality improvement program',
                'details': f'{quality_churn_pct:.1f}% of churn is due to service quality issues. Focus on network improvements and reliability.',
                'expected_impact': '25-35% reduction in quality-related churn'
            })

    # Customer Support
    if 'Customer Support' in category_analysis.index:
        support_churn_pct = (category_analysis['Customer Support'] / len(df_clean[df_clean[churn_col] == 'Yes'])) * 100
        if support_churn_pct > 5:
            recommendations.append({
                'priority': 'MEDIUM',
                'area': 'Customer Support',
                'recommendation': 'Enhance customer support experience and training',
                'details': f'{support_churn_pct:.1f}% of churn is due to customer support issues. Improve response times and resolution quality.',
                'expected_impact': '30-40% reduction in support-related churn'
            })
````

````python
# Contract-based recommendations
if 'Contract' in df_clean.columns and churn_col in df_clean.columns:
    month_to_month_churn = df_clean[df_clean['Contract'] == 'Month-to-month'][churn_col].apply(
        lambda x: x == 'Yes'
    ).mean()

    if month_to_month_churn > 0.4:  # If > 40% churn
        recommendations.append({
            'priority': 'HIGH',
            'area': 'Contract Strategy',
            'recommendation': 'Implement contract incentive programs',
            'details': f'Month-to-month customers have {month_to_month_churn:.1%} churn rate. Offer discounts for annual contracts.',
            'expected_impact': '15-25% churn reduction'
        })
````

````python
# Payment method recommendations
if 'Payment_Method' in df_clean.columns and churn_col in df_clean.columns:
    electronic_check_churn = df_clean[df_clean['Payment_Method'] == 'Electronic check'][churn_col].apply(
          lambda x: x == 'Yes'
    ).mean()

    if electronic_check_churn > 0.4:
        recommendations.append({
            'priority': 'HIGH',
            'area': 'Payment Experience',
            'recommendation': 'Improve electronic payment experience',
            'details': f'Electronic check users have {electronic_check_churn:.1%} churn rate. Enhance payment UX and offer auto-pay incentives.',
            'expected_impact': '10-15% churn reduction'
        })
````

````python
# Tenure-based recommendations
if 'Tenure_Category' in df_clean.columns and churn_col in df_clean.columns:
    new_customer_churn = df_clean[df_clean['Tenure_Category'] == '0-1 Year'][churn_col].apply(
        lambda x: x == 'Yes'
    ).mean()

    if new_customer_churn > 0.4:
        recommendations.append({
            'priority': 'HIGH',
            'area': 'Customer Onboarding',
            'recommendation': 'Enhance new customer onboarding program',
            'details': f'New customers (0-1 year) have {new_customer_churn:.1%} churn rate. Implement 90-day success program.',
            'expected_impact': '20-30% reduction in early churn'
        })
````

````python
# Service usage recommendations
if 'service_usage_score' in df_clean.columns and churn_col in df_clean.columns:
    low_usage_churn = df_clean[df_clean['service_usage_score'] <= 2][churn_col].apply(
        lambda x: x == 'Yes'
    ).mean()

    if low_usage_churn > 0.3:
        recommendations.append({
            'priority': 'MEDIUM',
            'area': 'Service Adoption',
            'recommendation': 'Increase service adoption through cross-selling',
            'details': f'Low-usage customers have {low_usage_churn:.1%} churn rate. Develop targeted upsell campaigns.',
            'expected_impact': '5-10% churn reduction'
        })
````

````python
# High-value customer recommendations
if 'customer_value_score' in df_clean.columns and churn_col in df_clean.columns:
    recommendations.append({
        'priority': 'HIGH',
        'area': 'Customer Success',
        'recommendation': 'Implement VIP customer success program',
        'details': 'Create dedicated success managers for high-value customers with proactive engagement.',
        'expected_impact': '25-35% improvement in retention'
        })
````

````python
# Print recommendations
for i, rec in enumerate(recommendations, 1):
    print(f"\n RECOMMENDATION {i} - {rec['priority']} PRIORITY")
    print(f"Area: {rec['area']}")
    print(f"Recommendation: {rec['recommendation']}")
    print(f"Details: {rec['details']}")
    print(f"Expected Impact: {rec['expected_impact']}")
````
